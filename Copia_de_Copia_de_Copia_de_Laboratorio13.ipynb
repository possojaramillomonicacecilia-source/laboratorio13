Prompt 1: Análisis de Sentimiento con TextBlob

Instrucciones:
Utiliza la librería TextBlob para analizar el sentimiento de las siguientes frases. Debes interpretar el sentimiento de las frases como positivo, negativo o neutral en base a su polaridad.

Frase 1: "The new compression algorithm significantly reduced latency, exceeding all performance expectations."
Frase 2: "The system experienced a critical failure during load testing due to an unstable memory leak."
Frase 3: "The database deployment was completed on Tuesday and the logs are being stored on the local server."

Objetivo:
Para cada frase, analiza la polaridad y asigna el sentimiento correspondiente:

Positivo si la polaridad es mayor que 0.
Negativo si la polaridad es menor que 0.
Neutral si la polaridad es 0.

```python
from textblob import TextBlob

frases = [
    "The new compression algorithm significantly reduced latency, exceeding all performance expectations.",
    "The system experienced a critical failure during load testing due to an unstable memory leak.",
    "The database deployment was completed on Tuesday and the logs are being stored on the local server."
]

for i, texto in enumerate(frases, 1):
    blob = TextBlob(texto)
    polaridad = blob.sentiment.polarity

    if polaridad > 0:
        sentimiento = "Positivo"
    elif polaridad < 0:
        sentimiento = "Negativo"
    else:
        sentimiento = "Neutral"

    print(f"Frase {i}: {sentimiento} (Polaridad: {polaridad:.2f})")
```

Frase 1: "The new compression algorithm significantly reduced latency, exceeding all performance expectation
Traducción:"El nuevo algoritmo de compresión redujo la latencia significativamente, superando todas las expectativas de rendimiento."
Frase 2: "The system experienced a critical failure during load testing due to an unstable memory leak."
Traducción:"El sistema experimentó un fallo crítico durante las pruebas de carga debido a una fuga de memoria inestable."
Frase 3: "The database deployment was completed on Tuesday and the logs are being stored on the local server."
Traducción:"El despliegue de la base de datos se completó el martes y los registros están siendo almacenados en el servidor local."

Prompt 2: Entrenamiento de Modelo de Clasificación de Sentimientos con Naive Bayes

Instrucciones:
Entrena un modelo de clasificación de sentimientos utilizando el algoritmo Naive Bayes. Utiliza los siguientes ejemplos de frases etiquetadas y evalúa la precisión de tu modelo.

Datos de ejemplo:

Frase 1: "This app is very intuitive!"
Frase 2: "I didn't enjoy the customer service."
Frase 3: "Quite average experience."
Frase 4: "Fantastic presentation!"
Frase 5: "The food was awful."

Objetivo:

Entrena un modelo de Naive Bayes con estas frases etiquetadas.
Evalúa la precisión del modelo utilizando las frases proporcionadas.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# 1. Datos
frases = [
    "This app is very intuitive!",
    "I didn't enjoy the customer service.",
    "Quite average experience.",
    "Fantastic presentation!",
    "The food was awful."
]
etiquetas = [1, 0, 2, 1, 0] # 1: Positivo, 0: Negativo, 2: Neutral

# 2. Vectorización (Convertir texto a matriz de números)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(frases)

# 3. Entrenamiento del modelo Naive Bayes
modelo = MultinomialNB()
modelo.fit(X, etiquetas)

# 4. Evaluación (Predicción sobre los mismos datos)
predicciones = modelo.predict(X)
precision = accuracy_score(etiquetas, predicciones)

print(f"Predicciones: {predicciones}")
print(f"Precisión del modelo: {precision * 100}%")
```
